```python
plt.figure(figsize=(50,12))
df = pd.read_csv("final.csv")
df["Date"] = pd.to_datetime(df["Date"], format="%Y%m%d%H%M%S")
df["OnlyDate"] = df["Date"].dt.date
df.groupby("OnlyDate")["crmapp_record_count"].sum()
daily_sum = df.groupby("Date")["armapp_record_count"].sum()
x_labels = daily_sum.index.strftime('%d-%b')
# Apply custom Y-axis formatting
def human_format(x, pos):
    if x >= 1_000_000:
        return f'{x/1_000_000:.0f}M'
    elif x >= 1_000:
        return f'{x/1_000:.0f}K'
    else:
        return str(int(x))
plt.gca().yaxis.set_major_formatter(FuncFormatter(human_format))
plt.plot(x_labels, daily_sum.values, marker='o')
plt.xlabel("Date (Day-Month)")
plt.ylabel("Total bcmapp_record_count")
plt.xticks(rotation=45)
plt.tight_layout()

# plt.plot(df["bcmapp_record_count"]);
# for date, group in ndf:
#     plt.plot(date,group["bcmapp_record_count"])
```
